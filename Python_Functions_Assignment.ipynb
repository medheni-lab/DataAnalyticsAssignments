**7. What are the advantages of using generators over regular functions?**

* Following are the advantages of generators over regular functions:

  * **Memory Efficiency:** As generators don't store etire result in the memory.
  * **Lazy Evaluation:** Generators compute values only when need.
  * **Can model infinite sequences:** They can built infinite data streams which are not possible for regular functions.
  * **Cleaner code for Iterators:** Generators are simple and elegant way to create custom iterators. No need to write _iter_() and _next_() manually.
  * **Faster startup Time:** As generators don't pre-calculate values, they often start executing faster than regular functions as the calculate entire result first.

```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

num_list = [10, 3, 5, 8, 2, 7] # Example list of numbers

result = sum_of_even_numbers(num_list) # Call the function

print("Sum of even numbers:", result) # Output the result

```

```python
#2. Create a Python function that accepts a string and returns the reverse of that string.

def reverse_string(s):
  return s[::-1]

text = "Hello"
reversed_text = reverse_string(text)
print("Reversed string:", reversed_text)
```

```python
#3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square_number(numbers):
  return [num **2 for num in numbers]

list = [1,2,3,4,5]
squared_list = square_number(list)
print("Squared list:", squared_list)
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

primes = [n for n in range(1, 201) if is_prime(n)]
print("Primes from 1 to 200:", primes)
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        self.count += 1
        result = self.a
        self.a, self.b = self.b, self.a + self.b
        return result

for num in FibonacciIterator(10):
    print(num, end=' ')
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exp):
    for i in range(max_exp + 1):
        yield 2 ** i

for power in powers_of_two(5):
    print(power)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_name):
    with open(file_name, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

data = [(1, 4), (2, 1), (3, 7), (4, 2)]
sorted_data = sorted(data, key=lambda x: x[1])
print("Sorted by second element:", sorted_data)

```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

celsius = [0, 20, 37, 100]
fahrenheit = list(map(lambda c: ((c * 9/5) + 32), celsius))
print(fahrenheit)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda ch: ch not in vowels, s))

text = "Hello, how are you?"
no_vowels = remove_vowels(text)
print("Without vowels:", no_vowels)
```
