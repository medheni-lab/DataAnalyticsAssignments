# **Theory Questions:**

**1. What is a random variable in probability theory?**

* A random variable is a variable that takes numerical values determined by the outcome of a random experiment. It maps outcomes of a sample space to real numbers.

**2. What are the types of random variables?**

* Discrete: Takes countable values (e.g., number of heads in 10 coin tosses).

* Continuous: Takes any value within a range (e.g., height, weight, time).

**3. What is the difference between discrete and continuous distributions?**

* Discrete distribution: Defined at specific values (e.g., binomial, Poisson).

* Continuous distribution: Defined over an interval, described with a density function (e.g., normal distribution).

**4. What are probability distribution functions (PDF)?**

* For continuous variables, a Probability Density Function (PDF) describes the likelihood of a variable taking a certain value. The probability is found over intervals using integrals.

**5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**

* PDF: Shows the density at a point (for continuous variables).

* CDF: Gives the probability that a variable is less than or equal to a specific value.

**6. What is a discrete uniform distribution?**

* A distribution where all outcomes are equally likely (e.g., rolling a fair die).

**7. What are the key properties of a Bernoulli distribution?**

* Two outcomes: success (1) or failure (0).

* One trial, with success probability
p, and failure 1-p.

**8. What is the binomial distribution, and how is it used in probability?**

* Models the number of successes in a fixed number of independent Bernoulli trials.
Used when there's a fixed number of identical trials, each with the same success probability.

**9. What is the Poisson distribution and where is it applied?**

* Models the number of times an event occurs in a fixed interval of time or space.
Used in contexts like call arrivals or traffic flow.

**10. What is a continuous uniform distribution?**
* A distribution where all outcomes in an interval[a,b] are equally likely.
PDF is constant: f(x)= 1/(b-a).

**11. What are the characteristics of a normal distribution?**

* Bell-shaped, symmetric around the mean.

* Defined by mean μ and standard deviationσ.

* Most values lie within ±1, ±2, and ±3 standard deviations of the mean.

**12. What is the standard normal distribution, and why is it important?**

* A normal distribution with μ=0 and σ=1.
Used to standardize variables and simplify calculations using Z-scores.

**13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**
* States that the sampling distribution of the sample mean approaches a normal distribution as sample size increases, regardless of the population's distribution.
It justifies using normal models in many inferential statistics.

**14. How does the Central Limit Theorem relate to the normal distribution?**
* It explains why normal distributions appear frequently in statistics—sample means tend to be normally distributed under certain conditions.

**15. What is the application of Z statistics in hypothesis testing?**
* Z-statistics test hypotheses about population means (when population standard deviation is known), using standard normal distribution.

**16. How do you calculate a Z-score, and what does it represent?**

* Z= (X-μ)/σ

Represents how many standard deviations a value X is from the mean μ.

**17. What are point estimates and interval estimates in statistics?**

* Point estimate: A single value estimate of a parameter (e.g., sample mean).

 Interval estimate: A range (e.g., confidence interval) likely to contain the parameter.

**18. What is the significance of confidence intervals in statistical analysis?**
They provide a range that likely contains the true population parameter, offering more information than a point estimate.

**19. What is the relationship between a Z-score and a confidence interval?**
* Z-scores determine the critical values used to calculate confidence intervals (e.g., ±1.96 for 95% confidence).

**20. How are Z-scores used to compare different distributions?**

* Z-scores standardize values from different distributions, allowing direct comparison regardless of original scales.

**21. What are the assumptions for applying the Central Limit Theorem?**

*
  * Samples are independent.

  * Identically distributed with finite mean and variance.

  * Sample size is sufficiently large (typically n≥30).

**22. What is the concept of expected value in a probability distribution?**

* The expected value (mean) is the long-run average outcome:
E[X]=∑xP(x) for discrete,
E[X]=∫xf(x)dx for continuous.

**23. How does a probability distribution relate to the expected outcome of a random variable?**

* The distribution determines how likely each outcome is, and the expected value summarizes these into a single measure of central tendency.
